```python
import sys
import pickle
import requests
from datetime import datetime
from bs4 import BeautifulSoup
```

```python
def get_tech_news(first_page, last_page, url):
    print("page number from {} to {}".format(first_page, last_page-1))
    data = []
    for page_number in range(first_page, last_page):
        print("page number: {}".format(page_number))
        data = data + get_page_news( url + 'page/' + str(page_number) + '/' )
    
    print('done')
    return(data)

def get_page_news(page_url):
    r = requests.get(page_url)
    r.encoding = "UTF-8"

    soup = BeautifulSoup(r.text, 'html.parser')
    news_blocks = soup.find_all(class_="entry-header")
    
    news = []
    for each_news in news_blocks:
        try:
            news_info = get_news_info(each_news)
        except:
            pass

        news.append(news_info)
    return(news)

def get_news_info(each_news):
    date  = get_date(each_news)
    title = get_title(each_news)
    link  = get_link(each_news)
    content = get_content(link)
    info = {'date' : date,
            'title': title,
            'link' : link,
            'content': content}
    return(info)

def get_date(news_block_node):
    if news_block_node.find(class_="entry-date published") != None :
        date_string = news_block_node.find(class_="entry-date published").get_text()
    else :
        date_string = news_block_node.find(class_="entry-date published updated").get_text()        
    return date_string
    
def get_title(news_block_node):
    return news_block_node.find(class_="entry-title").get_text()

def get_link(news_block_node):
    return news_block_node.find(class_="entry-title").a.get('href')

def get_content(link):
    r = requests.get(link)
    r.encoding = "UTF-8"
    soup = BeautifulSoup(r.text, 'html.parser')
    content_block = soup.find(class_="entry-content")
    p_nodes = content_block.find_all('p',limit=5)
    article = ''
    for p_node in p_nodes :
        article = article + p_node.get_text()
    return article.replace("\n", "")
```

```python
data = get_tech_news(first_page=1, last_page=101, url="https://buzzorange.com/techorange/tag/artificialintelligence/")
```

```python
sys.setrecursionlimit(100000)
with open('../data/techorange.pkl', 'wb') as handle:
    pickle.dump(data, handle)
```
